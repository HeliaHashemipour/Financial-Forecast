```python
price_change=[]
for index in df.index:
    if index==67205:
        break
    kal=df['Close'][index+1]-df['Close'][index]
    price_change.append(kal)
price_change[0]
```

```python
df['price_change']=np.zeros(67206)
df['price_change'][1:67206]=price_change
```

```python
df['Label']=np.zeros((67206))
for index1 in df.index:
    if index1==67205 :
        break
    elif df['price_change'][index1+1]>0:
        df['Label'][index1]=1
```

```python
df['Typical_Price']=(df['High']+df['Close']+df['Low'])/3
```

```python
MA_list=[]
for index3 in df.index:
    t=(df['Typical_Price'][index3:index3+20].sum())/20
    MA_list.append(t)
df['MA']=np.zeros((67206)) 
df['MA'][19:]=np.array(MA_list[0:67187])
df
```

```python
MD_list=[]
for index4 in range(0,67186):
    MD=0
    for i in df['Typical_Price'][index4:index4+20]:
        IU=abs((i-df['MA'][index4+20]))
        MD+=IU
    MD=MD/20
    MD_list.append(MD)
print(len(MD_list))
df['MD']=np.zeros((67206))
df['MD'][19:67205]=np.array(MD_list)
df.info()
```

```python
df['CCI']=np.zeros(67206)
df['CCI'][19:67205]=(df['Typical_Price'][19:67205]-df['MA'][19:67205])/(0.015*df['MD'][19:67205])
```

```python
df.info()
```

```python
RSI=[]
for index5 in df.index:
    sprofit=0
    sharm=0
    if index5==67186:
        break
    for j in df['price_change'][index5:index5+20]:
        if j>0 :
            sprofit+=j
        else:
            sharm+=j
    meanprofit=sprofit/20
    meanharm=abs(sharm/20)
    if df['price_change'][index5+19]>0:
        meannp=(meanprofit*19+ abs(df['price_change'][index5+19]))/20
        meannh=(meanharm*19+ 0)/20
    else:
        meannh=(meanharm*19+ abs(df['price_change'][index5+19]))/20
        meannp=(meanprofit*19+ 0)/20
    RS=meannp/meannh
    rsi=100-(100/(1+RS))
    RSI.append(rsi)
RSI[0]
```

```python
df['RSI']=np.zeros(67206)
df['RSI'][19:67205]=np.array(RSI)
```

```python
df.info()
```

```python
df['Tht/dollar']=df['Open']/df2['OPEN']
```

```python
price_change_dollar=[]
for index7 in df.index:
    if index7==67205:
        break
    ka=df2['OPEN'][index7+1]-df2['OPEN'][index7]
    price_change_dollar.append(ka)
df2['PRICE_CHANGE_dollar']=np.zeros(67206)
df2['PRICE_CHANGE_dollar'][1:67206]=np.array(price_change_dollar)
    
```

```python
RSI2=[]
for index6 in df2.index:
    sprofit2=0
    sharm2=0
    if index6==67186:
        break
    for j2 in df2['PRICE_CHANGE_dollar'][index6:index6+20]:
        if j2>0 :
            sprofit2+=j2
        else:
            sharm2+=j2
    meanprofit2=sprofit2/20
    meanharm2=abs(sharm2/20)
    if df2['PRICE_CHANGE_dollar'][index6+19]>0:
        meannp2=(meanprofit2*19+ abs(df2['PRICE_CHANGE_dollar'][index6+19]))/20
        meannh2=(meanharm2*19+ 0)/20
    else:
        meannh2=(meanharm2*19+ abs(df2['PRICE_CHANGE_dollar'][index6+19]))/20
        meannp2=(meanprofit2*19+ 0)/20
    RS2=meannp2/meannh2
    rsi2=100-(100/(1+RS2))
    RSI2.append(rsi2)
RSI2[0]
```

```python
df['RSI_DOLLAR']=np.zeros(67206)
df['RSI_DOLLAR'][19:67205]=np.array(RSI2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sclar=StandardScaler()
x=ndf[['Open', 'Close', 'High', 'Low',
       'price_change', 'Typical_Price', 'MA', 'MD', 'CCI', 'RSI',
       'Tht/dollar', 'RSI_DOLLAR']].values
y=ndf['Label'].values
x=sclar.fit_transform(x)
trax,tesx,tray,tesy=train_test_split(x,y,random_state=42,test_size=.33)
```

```python
from sklearn.tree import ExtraTreeClassifier
from sklearn.metrics import f1_score

treep=ExtraTreeClassifier(random_state=9)
treep.fit(trax,tray)
yhat=treep.predict(tesx)
print(f1_score(tesy,yhat))
```

```python
yhat2=treep.predict(x)
print(f1_score(y,yhat2))
```
